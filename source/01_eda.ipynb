```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming `merged_gdp` DataFrame is already defined and populated

# Extract relevant columns
gdp_ppp_columns = [col for col in merged_gdp.columns if 'GDP_PPP' in col]
gdp_exp_columns = [col for col in merged_gdp.columns if 'GDP_Education_Exp' in col]

# Extract years from column names for better labeling
gdp_ppp_years = [col.split('_')[-1] for col in gdp_ppp_columns]
gdp_exp_years = [col.split('_')[-1] for col in gdp_exp_columns]

# Create plots
fig, axes = plt.subplots(2, 1, figsize=(15, 12), sharex=True)

# Plot GDP_PPP
for country in merged_gdp['Reference area'].unique():
    country_data = merged_gdp[merged_gdp['Reference area'] == country]
    axes[0].plot(
        gdp_ppp_years,
        country_data[gdp_ppp_columns].values.flatten(),
        label=country
    )
axes[0].set_title("GDP (PPP) Over Time", fontsize=14)
axes[0].set_ylabel("GDP (PPP) per Capita (USD)", fontsize=12)
axes[0].tick_params(axis='x', labelrotation=45)
axes[0].legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=8)  # Place legend outside plot

# Plot GDP_Education_Exp
for country in merged_gdp['Reference area'].unique():
    country_data = merged_gdp[merged_gdp['Reference area'] == country]
    axes[1].plot(
        gdp_exp_years,
        country_data[gdp_exp_columns].values.flatten(),
        label=country
    )
axes[1].set_title("GDP Education Expenditure Over Time", fontsize=14)
axes[1].set_xlabel("Years", fontsize=12)
axes[1].set_ylabel("Education Expenditure (% of GDP)", fontsize=12)
axes[1].tick_params(axis='x', labelrotation=45)
axes[1].legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=8)  # Place legend outside plot

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot_education_hours_mpl(df, selected_countries=None, figsize=(12, 15)):
    """
    Create three subplots showing education hours trends for primary, lower secondary,
    and upper secondary education using matplotlib.
    """
    # Extract years from column names
    def get_years(cols, level):
        return sorted([int(col.split('_')[-2]) for col in cols if level in col])
    
    # Get columns for each education level
    lower_sec_cols = [col for col in df.columns if 'lower_secondary_general' in col]
    upper_sec_cols = [col for col in df.columns if 'upper_secondary_general' in col]
    primary_cols = [col for col in df.columns if 'primary' in col and 'lower' not in col]
    
    # Get unique years for each level
    lower_years = get_years(lower_sec_cols, 'lower')
    upper_years = get_years(upper_sec_cols, 'upper')
    primary_years = get_years(primary_cols, 'primary')
    
    # If no countries specified, select top 10 by most recent year
    if selected_countries is None:
        latest_primary_col = [col for col in primary_cols if str(max(primary_years)) in col][0]
        selected_countries = df.nlargest(10, latest_primary_col)['Country'].tolist()
    
    # Create figure and subplots
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=figsize)
    
    # Define color cycle
    colors = plt.cm.tab10(np.linspace(0, 1, len(selected_countries)))
    
    # Plot each country
    for country, color in zip(selected_countries, colors):
        country_data = df[df['Country'] == country]
        
        if not country_data.empty:
            # Primary education
            values = [country_data[f'primary_{year}_hours'].iloc[0] 
                     if f'primary_{year}_hours' in df.columns else np.nan 
                     for year in primary_years]
            ax1.plot(primary_years, values, marker='o', label=country, color=color)
            
            # Lower secondary education
            values = [country_data[f'lower_secondary_general_{year}_hours'].iloc[0] 
                     if f'lower_secondary_general_{year}_hours' in df.columns else np.nan 
                     for year in lower_years]
            ax2.plot(lower_years, values, marker='o', color=color)
            
            # Upper secondary education
            values = [country_data[f'upper_secondary_general_{year}_hours'].iloc[0] 
                     if f'upper_secondary_general_{year}_hours' in df.columns else np.nan 
                     for year in upper_years]
            ax3.plot(upper_years, values, marker='o', color=color)
    
    # Customize subplots
    for ax, title, years in zip(
        [ax1, ax2, ax3],
        ['Primary Education Hours', 
         'Lower Secondary Education Hours', 
         'Upper Secondary Education Hours'],
        [primary_years, lower_years, upper_years]
    ):
        ax.set_title(title, pad=15)
        ax.set_ylabel('Hours per Year')
        ax.grid(True, linestyle='--', alpha=0.7)
        ax.set_xticks(years)
        ax.set_xticklabels(years, rotation=45)
    
    # Add legend only to first subplot
    ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # Add xlabel to bottom subplot
    ax3.set_xlabel('Year')
    
    # Add main title
    fig.suptitle('Annual Education Hours by Country and Level', 
                 fontsize=16, y=0.95)
    
    # Adjust layout
    plt.tight_layout()
    
    return fig, (ax1, ax2, ax3)
```

```python
# Para ver los top 10 países
fig, axes = plot_education_hours_mpl(teaching_time)
plt.show()
```

```python
def merge_education_dataframes(df_hours, df_ratio):
    # Create copies to avoid modifying original dataframes
    df_hours = df_hours.copy()
    df_ratio = df_ratio.copy()
    
    # Add prefix to all columns except 'Country'
    for df, prefix in [(df_hours, 'teaching_hours_'), (df_ratio, 'student_teacher_ratio_')]:
        # Get all columns except 'Country'
        cols_to_rename = [col for col in df.columns if col != 'Country']
        
        # Create rename mapping
        rename_dict = {col: f"{prefix}{col}" for col in cols_to_rename}
        
        # Rename columns
        df.rename(columns=rename_dict, inplace=True)
    
    # Merge the dataframes on the country column
    merged_df = pd.merge(
        df_hours,
        df_ratio,
        on='Country',
        how='outer'  # Use outer join to keep all countries from both dataframes
    )
    
    # Sort by country name
    merged_df = merged_df.sort_values('Country')
    
    # Reset the index
    merged_df = merged_df.reset_index(drop=True)
    
    return merged_df

# Create merged dataframe
merged_education = merge_education_dataframes(teaching_time, teaching_staff_ratio)
```
