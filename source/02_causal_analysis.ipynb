```python
VAR_MAP = {
    'gdp': 'GDP_PPP_',
    'edu_exp': 'GDP_Education_Exp_',
    'ict': 'ict_access_pct_pop_',
    'gini': 'gini_coef_',
    'math': 'math_',
    'reading': 'reading_',
    'science': 'science_',
    'hours': 'teaching_hours_',
    'ratio': 'student_teacher_ratio_',
}

PISA_VARS = {'math', 'reading', 'science'}
```

```python
# Inicializar un diccionario para almacenar los DataFrames
dataframes = {}

# Iterar sobre cada variable y su prefijo en VAR_MAP
for var, prefix in VAR_MAP.items():
    # Seleccionar las columnas que empiezan con el prefijo correspondiente
    cols = ['Country'] + [col for col in data.columns if col.startswith(prefix)]
    
    # Verificar si existen columnas que coincidan con el prefijo
    if len(cols) > 1:
        # Crear el DataFrame para la variable actual
        dataframes[var] = data[cols].copy()
    else:
        print(f"No se encontraron columnas para la variable '{var}' con el prefijo '{prefix}'.")

# Opcional: Mostrar las claves del diccionario para verificar
print(dataframes.keys())
```

```python
dataframes['treatment'] = data[['Country', 'Treatment']].copy()
dataframes['deaths'] = data[['Country', 'Pandemic_Deaths_2020']].copy()
dataframes['death_pct'] = data[['Country', 'COVID_Death_Percent_Population']].copy()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the variable mappings and labels
VAR_MAP = {
    'gdp': 'GDP_PPP_',
    'edu_exp': 'GDP_Education_Exp_',
    'ict': 'ict_access_pct_pop_',
    'gini': 'gini_coef_',
    'math': 'math_',
    'reading': 'reading_',
    'science': 'science_',
    'hours': 'teaching_hours_',
    'ratio': 'student_teacher_ratio_'
}

PISA_VARS = {'math', 'reading', 'science'}

PLOT_LABELS = {
    'gdp': 'GDP PPP',
    'edu_exp': 'Education Expenditure (% of GDP)',
    'ict': 'ICT Access (%)',
    'gini': 'Gini Coefficient',
    'math': 'Math PISA Scores',
    'reading': 'Reading PISA Scores',
    'science': 'Science PISA Scores',
    'hours': 'Teaching Hours',
    'ratio': 'Student-Teacher Ratio'
}

def plot_variable_timeseries(df, countries, var, start_year=None, end_year=None, 
                             plot=True, figsize=(12, 8), include_errors=False):
    """
    Plot the timeseries of a specified variable for a list of countries.

    Parameters:
    -----------
    df : pandas DataFrame
        The input dataframe containing all data.
    countries : list of str
        List of country names to plot.
    var : str
        Variable name from VAR_MAP (e.g., 'gdp', 'ict', 'gini', 'math', etc.).
    start_year : int, optional
        Start year for the timeseries (applicable to regular variables).
    end_year : int, optional
        End year for the timeseries (applicable to regular variables).
    plot : bool, optional
        Whether to display the plot (default: True).
    figsize : tuple, optional
        Figure size for the plot (default: (12, 8)).
    include_errors : bool, optional
        Whether to include error bars for PISA variables (default: False).

    Returns:
    --------
    pandas DataFrame
        A dataframe containing the timeseries data for all specified countries and the given variable.
    """
    # Validate variable
    if var not in VAR_MAP:
        raise ValueError(f"Variable '{var}' not found. Available variables: {list(VAR_MAP.keys())}")
    
    prefix = VAR_MAP[var]
    is_pisa = var in PISA_VARS
    
    # Initialize list to collect DataFrames
    data_list = []
    
    # Initialize plot
    if plot:
        plt.figure(figsize=figsize)
        sns.set_style("whitegrid")
    
    for country in countries:
        try:
            # Filter data for the current country
            data_row = df[df['Country'] == country]
            if data_row.empty:
                print(f"Country '{country}' not found in the dataset. Skipping.")
                continue
            
            if is_pisa:
                # Handle PISA variables with Mean and SE
                mean_cols = [col for col in df.columns if col.startswith(prefix) and 'Mean' in col]
                se_cols = [col for col in df.columns if col.startswith(prefix) and 'SE' in col]
                
                # Extract years from column names and convert to integers
                years = sorted([int(col.split('_')[1]) for col in mean_cols if col.split('_')[1].isdigit()])
                
                # Apply year filters if specified
                if start_year is not None:
                    years = [year for year in years if year >= start_year]
                if end_year is not None:
                    years = [year for year in years if year <= end_year]
                
                means = []
                errors = []
                valid_years = []
                
                for year in years:
                    mean_col = f"{prefix}{year}_Mean"
                    se_col = f"{prefix}{year}_SE"
                    
                    if mean_col in df.columns and se_col in df.columns:
                        mean_val = data_row.iloc[0][mean_col]
                        se_val = data_row.iloc[0][se_col]
                        
                        if pd.notna(mean_val) and pd.notna(se_val):
                            means.append(float(mean_val))
                            errors.append(float(se_val))
                            valid_years.append(year)
                
                if not valid_years:
                    print(f"No valid data for {country} and variable '{var}'. Skipping.")
                    continue
                
                # Append to data list
                temp_df = pd.DataFrame({
                    'country': country,
                    'year': valid_years,
                    'mean': means,
                    'se': errors
                })
                data_list.append(temp_df)
                
                if plot:
                    label = country
                    plt.errorbar(valid_years, means, yerr=errors if include_errors else None,
                                 fmt='o-', capsize=5, capthick=1.0,
                                 linewidth=2, markersize=6, label=label)
            
            else:
                # Handle regular variables
                var_cols = [col for col in df.columns if col.startswith(prefix) and col.split('_')[-1].isdigit()]
                years = [int(col.split('_')[-1]) for col in var_cols if col.split('_')[-1].isdigit()]
                
                # Sort the columns based on years
                sorted_pairs = sorted(zip(years, var_cols))
                sorted_var_cols = [col for year, col in sorted_pairs]
                sorted_years = [year for year, col in sorted_pairs]
                
                # Apply year filters if specified
                if start_year is not None:
                    sorted_var_cols = [col for col, year in zip(sorted_var_cols, sorted_years) if year >= start_year]
                    sorted_years = [year for year in sorted_years if year >= start_year]
                if end_year is not None:
                    sorted_var_cols = [col for col, year in zip(sorted_var_cols, sorted_years) if year <= end_year]
                    sorted_years = [year for year in sorted_years if year <= end_year]
                
                values = data_row[sorted_var_cols].iloc[0].astype(float)
                
                # Handle missing data
                valid_mask = values.notna()
                valid_years = [year for year, valid in zip(sorted_years, valid_mask) if valid]
                valid_values = [val for val, valid in zip(values, valid_mask) if valid]
                
                if not valid_years:
                    print(f"No valid data for {country} and variable '{var}'. Skipping.")
                    continue
                
                # Append to data list
                temp_df = pd.DataFrame({
                    'country': country,
                    'year': valid_years,
                    'value': valid_values
                })
                data_list.append(temp_df)
                
                if plot:
                    label = country
                    plt.plot(valid_years, valid_values, marker='o', linewidth=2, markersize=6, label=label)
                    # Assuming regular variables do not have SE
                
        except Exception as e:
            print(f"Error processing {country} for variable '{var}': {e}")
            continue
    
    if plot:
        plt.xlabel('Year')
        plt.ylabel(PLOT_LABELS.get(var, var))
        plt.title(f'{PLOT_LABELS.get(var, var)} Timeseries for Selected Countries')
        plt.legend(title='Country')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    
    if not data_list:
        print("No data available to return.")
        return pd.DataFrame()
    
    # Concatenate all collected data
    combined_df = pd.concat(data_list, ignore_index=True)
    
    if is_pisa:
        pivot_df = combined_df.pivot(index='year', columns='country', values=['mean', 'se'])
        # Flatten the MultiIndex columns
        pivot_df.columns = ['_'.join(col).strip() for col in pivot_df.columns.values]
        return pivot_df
    else:
        pivot_df = combined_df.pivot(index='year', columns='country', values='value')
        return pivot_df
```
