```python
import pandas as pd
```

```python
data = pd.read_csv("../data/aggregated_df.csv")
data
```

```python
VAR_MAP = {
    'gdp': 'GDP_PPP_',
    'edu_exp': 'GDP_Education_Exp_',
    'ict': 'ict_access_pct_pop_',
    'gini': 'gini_coef_',
    'math': 'math_',
    'reading': 'reading_',
    'science': 'science_',
    'hours': 'teaching_hours_',
    'ratio': 'student_teacher_ratio_'
}

PISA_VARS = {'math', 'reading', 'science'}
```

```python
# Inicializar un diccionario para almacenar los DataFrames
dataframes = {}

# Iterar sobre cada variable y su prefijo en VAR_MAP
for var, prefix in VAR_MAP.items():
    # Seleccionar las columnas que empiezan con el prefijo correspondiente
    cols = ['Country'] + [col for col in data.columns if col.startswith(prefix)]
    
    # Verificar si existen columnas que coincidan con el prefijo
    if len(cols) > 1:
        # Crear el DataFrame para la variable actual
        dataframes[var] = data[cols].copy()
    else:
        print(f"No se encontraron columnas para la variable '{var}' con el prefijo '{prefix}'.")

# Opcional: Mostrar las claves del diccionario para verificar
print(dataframes.keys())
```
